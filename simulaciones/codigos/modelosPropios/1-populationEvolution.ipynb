# Evolución de una población de una especie

### Descripción del modelo

Modelo tipo "no espacial" similar al planteado en el PRE de McKane de 2004.
El sistema cuenta con $N$ sitios, los cuales están etiquetados con:
- La letra $A$ si se encuentran ocupados por la especie en cuestión. En cada instante, la cantidad de sitios de este tipo lo denotaremos con la cantidad $n$.
- La letra $E$ si se encuentra vacío. Naturalmente, habrá $N-n$ sitios de este tipo.

### Evolución del modelo

##### Descripción de la evolución

El modelo evoluciona según el siguiente proceso estocástico con reglas de interacción del 

1. Con probabilidad $u$ se elige un sitio $S_1$ al azar.
    - Proceso de muerte: Si tal sitio es $S_1 = A$ (se encuentra ocupado), entonces se da la transición $A \rightarrow E$ con una probabilidad $p_\mu$.
    - Proceso de migración expontánea: Si tal sitio es $S_1 = E$ (se encuentra desocupado), entonces se da la transición $E \rightarrow A$ con una probabilidad $p_\alpha$.

2. En otro caso (con probabilidad $1-u$) se elige un sitio $S_2$ al azar:
    - Proceso de nacimiento: Si tal sitio es $S_2 = A$ (se encuentra ocupado) se elije otro sitio, $S_3$, al azar
        - Si ese sitio es $S_3 = E$ (se encuentra desocupado) entonces se da la transición $AE \rightarrow AA$ con probabilidad $p_\nu$.
        - Si ese sitio es $S_3 = A$ (se encuentra ocupado) entonces no se da ninguna transición y el proceso termina
    - En el caso que el sitio $S_2 = E$ (se encuentra ocupado), el proceso se termina.

#### Tasas de transiciones



El estado del sistema viene caracterizado por el número $n(t)$ a lo largo de los pasos del tiempo. La tasa de transición de un estado $n(t)$ a un estado $n'(t+1)$ en un paso de tiempo será denotado como $T(n'|n)$. De esta manera, como en cada paso de tiempo puede incrementearse en uno o disminuir en uno la cantiadad $n$, las tasas de transiciones no nulas son

- $T(n+1|n)$: debido a procesos de nacimiento y de migración expontánea.
- $T(n-1|n)$: debido a procesos de muerte.

Las expresiones son, entonces,

$$T(n+1|n) = up_\alpha \frac{N-n}{N} + (1-u) p_\nu \frac{n}{N}\frac{N-n}{N-1},$$
$$T(n-1|n) = u p_\mu \frac{n}{N}.$$

#### Ecuación Maestra y evolución del valor medio.

Sea $P(n, t)$ la probabilidad de encontrar a tiempo $t$ el sistema con $n$ sitios etiquetados con $A$. La ecuación Maestra que gobierna su evolución en el tiempo es

$$\frac{dP(n, t)}{dt} = T(n|n+1) P(n+1,t) + T(n|n-1)P(n-1,t) - \left[T(n+1|n) + T(n-1|n)\right] P(n, t).$$

Al evaluar la cantidad $\left\langle n \right\rangle = \sum_{n=0}^{N} n P(n, t)$ y tomar su derivada temporal $\frac{d\left\langle n \right\rangle}{dt} =  \sum_{n=0}^{N} n \frac{dP(n, t)}{dt}$, resulta

$$\frac{d\left\langle n \right\rangle}{dt} = \sum_{n=0}^{N} n \left\{ T(n|n+1) P(n+1,t) + T(n|n-1)P(n-1,t) - \left[T(n+1|n) + T(n-1|n)\right] P(n, t) \right\}$$

$$\Rightarrow \frac{d\left\langle n \right\rangle}{dt} = \left\{\sum_{n = 1}^{N+1} (n-1) T(n-1|n) P(n, t) - \sum_{n = 0}^{N} n T(n-1|n) P(n, t) \right\} +$$

$$+ \left\{\sum_{n = -1}^{N-1} (n+1) T(n+1|n) P(n, t) - \sum_{n = 0}^{N} n T(n+1|n) P(n, t) \right\}$$

Tomamos $T(N|N+1) = T(N+1|N) = T(-1|0) = 0$, entonces podemos compactar las expresiones anteriores directamente 

$$\frac{d\left\langle n \right\rangle}{dt} = \sum_{n = 0}^{N} \left\{T(n+1|n) - T(n-1|n)\right\} P(n, t).$$


Reemplazando por las expresiones de las tasas de transición obtenidas más arriba, arribamos directamente a

$$\frac{d\left\langle n \right\rangle / N}{dt} = \frac{up_\alpha}{N} \left\langle 1-\frac{n}{N}\right\rangle + \frac{(1-u)p_\nu}{N-1} \left\langle \frac{n}{N} \left(1-\frac{n}{N}\right)\right\rangle - \frac{up_\mu}{N} \left\langle \frac{n}{N} \right\rangle.$$

#### Valor estacionario.

Denominando $\phi(t) = \frac{\langle n (t) \rangle}{N}$, al considerar la aproximación de campo medio ($\langle n^2 \rangle \approx  \langle n \rangle^2$) la ecuación diferencia que gobierna la evolución de la densidad $\phi$ es

$$\frac{d\phi}{dt} = \frac{up_\alpha}{N} + \left(\frac{(1-u)p_\nu}{N-1}  - \frac{u(p_\alpha + p_\mu)}{N} \right) \phi - \frac{(1-u) p_\nu}{N-1} \phi^2$$

Podemos preguntarnos la existencia de un equilibrio de este sistema. En tal caso, $\frac{d\phi}{dt} = 0$, de modo que

$$ 0 = \phi_{est}^2 + \left(\frac{u}{1-u} \frac{N-1}{N} \frac{p_\alpha + p_\mu}{p_\nu} - 1\right) \phi_{est} - \frac{u}{1-u}\frac{N-1}{N} \frac{p_{\alpha}}{p_\nu}$$

Sea $K = \frac{u}{1-u} \frac{N-1}{N} \frac{p_\mu}{p_\nu}$, de modo que

$$0 = \phi_{est}^2 + \left(K(1+\frac{p_\alpha}{p_\mu}) - 1\right) \phi_{est} - K\frac{p_\alpha}{p_\mu}$$

$$\Rightarrow \phi_{est} = - \frac{1}{2}\left(K(1+\frac{p_\alpha}{p_\mu}) - 1\right) + \frac{1}{2} \sqrt{\left(K(1+\frac{p_\alpha}{p_\mu}) - 1\right)^2 + 4K\frac{p_\alpha}{p_\mu}},$$

pues $\phi_{est} > 0$.

Acá se puede hacer un Taylor con $\frac{p_\alpha}{p_\nu} \ll 1$.

### Simulación numérica

##### Código de la simulación

Haremos una simulación numérica en `Python`. Para las simulaciones, tomaremos $P(n, 0) = \delta_{n, 0}$. 

```python
# imoprt libraries
import random
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
```

```python
# Determine the specie choiced
def specie_choiced(place, n):
    if place < n:
        return "A"
    else:
        return "E"

# Choice of the species in the places
def species_choiced(place1, place2, n):
    specie1 = specie_choiced(place1, n)
    specie2 = specie_choiced(place2, n)
    return specie1, specie2
```

```python
# Population evolution
def population_evolution(N, n_initial, u, pmu, pnu, palpha, t_max):
    #Initialization of n
    n = n_initial

    #Initialization of the list of n
    history_n = [n]

    #Iterate over time
    for t in range(t_max):
        #Choice he the interaction (u: single, 1-u: double)
        if random.random() < u:
            #Single interaction
            place = random.randint(0, N-1)
            specie = specie_choiced(place, n)
            if specie == "A": # The place choose is A
                if random.random() < pmu:
                    # Dying proccess occurs (A -> E)
                    n = n - 1   # "A" site is converted to "E"
            else: # The place choose is E
                if random.random() < palpha:
                    # Expontanical migration proccess occurs (E -> A)
                    n = n + 1   # "E" site is converted to "A"
                    
        else:
            #Double interaction
            place1, place2 = random.sample(range(N), 2)
            specie1, specie2 = species_choiced(place1, place2, n)
            if specie1 == "A" and specie2 == "E": 
                # First place is A and second place is E
                if random.random() < pnu:
                    # Birth proccess occurs (A + E -> A + A)
                    n = n + 1

            # Any other combination, nothing happens
        
        #Append the new value of n to the history
        history_n.append(n)
    
    return history_n
```

#### Primera simulación

Comenzamos analizando nuestro sistema para $N = 20$ y $p_\mu=0.2$. Además, $p_\alpha/p_\mu = 0.01$ y $p_\nu/p_\mu=2$. Para todas las simulaciones, a menos que se diga lo contrario, utilizaremos $u = 0.5$. Número de itraciones $t_{max} = 100000$

```python
# Parameters
u = 0.5         # Single interaction probability
N = 20          # Number of sites
n_initial = 0   # Initial number of "A" sites
pmu = 0.2       # Death probability
pnu = 2*pmu     # Birth probability
palpha = 0.01*pmu # Migration probability
t_max = 100000   # Number of iterations
```

```python
# Simulation 
history_n = population_evolution(N, n_initial, u, pmu, pnu, palpha, t_max)
```

```python
# figures

fig = plt.figure(figsize=(12, 5))

# plot the density of A sites
ax = plt.subplot2grid((1, 4), (0, 0), colspan=3)
ax.plot(np.array(history_n)/N, lw=1, color='blue')
ax.set_xlabel(r"$t$")
ax.set_ylabel(r"$\phi$")
ax.set_title(r"Density of $A$ sites. $N = 20$")
ax.set_yticks([0, 0.2, 0.4, 0.6, 0.8, 1])
ax.set_ylim([0, 1])

# plot the histogram
ax = plt.subplot2grid((1, 4), (0, 3))
ax.hist(np.array(history_n)/N, bins = np.linspace(0, 1, N+2), orientation='horizontal', align = "left", rwidth = 0.75, color='blue')
ax.set_xlabel(r"Number of times")
ax.set_title(r"Histogram of $n$")

# do not show the numbers in the y-axis
ax.set_yticks([])
ax.set_yticklabels([])
ax.set_ylim([0, 1])

plt.tight_layout()

plt.show()
```

#### Ensamble de simulaciones

Haremos un ensamble de varias evoluciones con las mismas condiciones iniciales. Para ello, haremos un ensamble inicial de $n_{runs} = 500$. Veremos el efecto que tenga promediar sobre varias evoluciones.

```python
def ensamble_population_evolution(N, n_initial, u, pmu, pnu, palpha, t_max, n_runs):
    #Initialization of the list of n
    history_n = np.zeros((n_runs, t_max+1))
    history_n[:, 0] = n_initial

    #Iterate over runs
    for i in tqdm(range(n_runs)):
        history_n[i, :] = population_evolution(N, n_initial, u, pmu, pnu, palpha, t_max)

    return history_n
```

```python
# Observables
def get_density_from_ensamble(history_n, N):
    return np.mean(history_n, axis = 0)/N
```

```python
# Same parameters and initial conditions
n_runs = 500
history_n = ensamble_population_evolution(N, n_initial, u, pmu, pnu, palpha, t_max, n_runs)
```

```python
density = get_density_from_ensamble(history_n, N)
```

```python
# Theoretical density prediction

def theoretical_prediction(u, N, pmu, pnu, palpha):
    K = u/(1-u) * (N-1)/N * pmu/pnu
    phi_est = -0.5*(K*(1+palpha/pmu) - 1) + 0.5*np.sqrt((K*(1+palpha/pmu) - 1)**2 + 4*K*palpha/pmu)
    return phi_est

phi_est = theoretical_prediction(u, N, pmu, pnu, palpha)
```

```python
# Plots
fig = plt.figure(figsize=(12, 5))

# plot the density of A sites
ax = plt.subplot2grid((1, 4), (0, 0), colspan=3)
ax.plot(density, lw=1, color='blue')
ax.axhline(y=phi_est, color='red', linestyle='--')
ax.set_xlabel(r"$t$")
ax.set_ylabel(r"$\phi$")
ax.set_title(r"Density of $A$ sites. $N = 20$")
ax.set_yticks([0, 0.2, 0.4, 0.6, 0.8, 1])
ax.set_ylim([0, 1])
ax.legend([r"Ensamble simulations", r"Theoretical prediction"])


# plot the histogram
ax = plt.subplot2grid((1, 4), (0, 3))
ax.hist(density, bins = np.linspace(0, 1, N+2), orientation='horizontal', align = "left", rwidth = 0.75, color='blue')
ax.set_xlabel(r"Number of times")
ax.set_title(r"Histogram of $n$")
ax.set_yticks([])
ax.set_yticklabels([])
ax.set_ylim([0, 1])

plt.tight_layout()

plt.show()
```

Dos observaciones:
- Al promediar sobre un ensamble, desaparecen las oscilaciones. El histograma ahora adopta una distribución más angosta y condensada al rededor del valor más problable.
- Al contrastar con el valor de equilibrio, este se encuentra por encima de la distribución encontrada en el ensamble de las simulaciones. Por ahora no entiendo qué es lo que está pasando. Quizás dependa de las condiciones iniciales. EDIT: El valor de equilibrio hallado corresponde a haber hecho una aproximación de campo medio al haber tomado $\langle n^2 \rangle \approx \langle n \rangle^2$, que en principio debería ocurrir para $N\rightarrow \infty$

#### Cuantificación del proceso en función de la cantidad de veces que la población se extingue, $n_{ext}$

Para identificar la cantidad de veces que la población se extingue, lo haremos en corridas de $t_{max} = 100000$. Haremos ensambles de $t_{runs} = 100$. Notar que en este caso no nos interesa la historia completa, sólo la cantidad de veces que el sistema se extingue. Tendremos un contador que acumule esta información. 

```python
# counter of the number of times that the density becomes zero (extinction)
def extinction_counter(N, n_initial, u, pmu, pnu, palpha, t_max, t_runs):
    counters = []
    for i in range(t_runs):
        counter = 0
        # We are not interested in the history of the population
        n = n_initial

        #Iterate over time
        for t in range(t_max):
            # Choice he the interaction (u: single, 1-u: double)
            if random.random() < u:
                # Single interaction
                place = random.randint(0, N-1)
                specie = specie_choiced(place, n)
                if specie == "A":
                    if random.random() < pmu:
                        # Dying proccess occurs (A -> E)
                        n = n - 1
                        # If the population becomes zero, we count it
                        if n == 0:
                            counter += 1
                else:
                    if random.random() < palpha:
                        # Expontanical migration proccess occurs (E -> A)
                        n = n + 1
            else:
                # Double interaction
                place1, place2 = random.sample(range(N), 2)
                specie1, specie2 = species_choiced(place1, place2, n)
                if specie1 == "A" and specie2 == "E":
                    if random.random() < pnu:
                        # Birth proccess occurs (A + E -> A + A)
                        n = n + 1
                # Any other combination, nothing happens

        counters.append(counter)
    return counters
```

```python
# Parameters
N = 20
pmu = 0.2
pnus = np.array([0.25, 0.5, 0.75, 1, 1.2, 1.4, 1.6, 1.8, 2, 2.25, 2.5, 2.75, 3, 3.5, 4])*pmu
palpha = 0.01*pmu
t_max = 100000
t_runs = 100
```

```python
# Simulations
extinctions = [extinction_counter(N, n_initial, u, pmu, pnu, palpha, t_max, t_runs) for pnu in tqdm(pnus)]
```

Comentario: Simulación mucho más costosa por la cantidad de $t_{runs}$

```python
# Plot
fig = plt.figure(figsize=(10, 5))

plt.errorbar(pnus/pmu, np.mean(extinctions, axis = 1), yerr = np.std(extinctions, axis = 1), fmt='o', color='blue')
plt.xlabel(r"$p_\nu/p_\mu$")
plt.ylabel(r"Extinctions for each $t_{max} = 100000$")
plt.title(r"Extinctions")
plt.legend([r"Ensamble simulations"])

plt.show()
```
